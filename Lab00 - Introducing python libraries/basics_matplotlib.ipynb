```python
#plotagem da função exponencial de x no intervalo entre 1 e 5 usando
#matplotlib e numpy

import numpy as np
import matplotlib.pyplot as plt

x = np.arange(1.0, 5.1, 0.2) #gera um array numpy onde elementos seguem uma PA
#que vai de 1 até 5.1 com razão 0.2
y = np.exp(x)  #gera um array numpy com os valores dos exponenciais de cada
#elemento em x

plt.plot(x, y, 'bo',  color='red') #o terceiro argumento controla como os pontos são plotados
#algumas possíveis opções para esse terceiro argumento são:
#'b-': linha cheia na cor azul
#'r--': linha tracejada na cor vermelha
#'y*': coloca os pontos com um asterisco em amarelo
#'bo': coloca os pontos com uma bolinha azul

plt.title('Meu gráfico de exponencial!') #coloca um título

plt.text( 1, 100, "Texto no ponto 1,100" ) #coloca um texto no gráfico

plt.show()

```

```python
#plotando várias curvas na mesma janela de plot com matplotlib
#vamoplotar 1.2^x, log(x) e cos(x)


x = np.arange(1.0, 5.1, 0.1)
y = 1.2 ** x

x2 = x[:]  #gera cópia raza de x
y2 = np.log(x2)

x3 = x
y3 = np.cos(x3)

plt.title('Três funções')

plt.plot(x, y, 'go', label='Pontos azuis')  # 'bo' significa pontos (o) azuis (b)
plt.plot(x2, y2, color='pink', label='Linha rosa')  # linha rosa
plt.plot(x3, y3, 'r--', label='Linha vermelha tracejada')  # 'r--' significa linha vermelha (r) tracejada (--)


plt.legend( ('Exponencial', 'Logaritmo', 'Cosseno'), loc='lower right' )

plt.show()
```

```python
#exemplo: exibição de três gráficos em separado

import matplotlib.pyplot as plt
import numpy as np

#vamos usar o conceito de subplot

#indica que temos uma tabela de plotagem com uma linha, 3 colunas e
#estamos trabalhando na subplotagem 1
plt.subplot(1, 3, 1)

pi = np.pi

x1 = np.arange( -4*pi, 4*pi, 0.1)
y1 = np.sin(x1)
plt.title('seno')
plt.plot(x1, y1, 'b--')

#agora, estamos trabalhando na subplotagem 2
plt.subplot(1, 3, 2)
x2 = x1
y2 = np.cos(x2)
plt.title('cosseno')
plt.plot(x2, y2, 'ro')

#agora, estamos trabalhando na subplotagem 3
plt.subplot(1, 3, 3)
x3 = x1
y3 = np.tan(x3)
plt.plot(x3, y3, 'y')
plt.title('tangente')
plt.show()
```

```python
#exemplo de código que salva o gráfico de x^3 + 4x + 2

fig = plt.figure() #retorna um objeto que manuseia a figura corrente

x = np.arange(-10, 10.1, 0.1)
y = x**3 + 4*x + 2

plt.plot( x, y )

#define a janela de plotagem: o eixo x vai de -12 até 15 e o eixo y
#vai de -1000 até 1000
plt.axis( [-12, 15, -1000, 1000] )

plt.xlabel( 'meu eixo x' ) #dando um label (etiqueta) para o eixo x
plt.ylabel( '\nmeu eixo y' ) #dando um label (etiqueta) para o eixo y

fig.savefig( R'.\grafico.png') # An 'r' before a string tells the Python interpreter to treat backslashes as a literal (raw) character

plt.show()



```

```python
#exemplo: gráfico de barras com matplotlib

import matplotlib.pyplot as plt

#coordenadas dos cantos inferiores de cada barra
left = [1, 2, 3, 4]
height = [42, 21, 16, 4] #altura das barras

estados = ('SP', 'MG', 'RJ', 'ES')

plt.bar(left, height, color='purple', tick_label=estados )

plt.title('População dos estados do sudeste')
plt.xlabel('Estado')
plt.ylabel('Milhões de habitantes')

plt.show()
```

```python
#exemplo: plotando um histograma com a probabilidade de pontuação de QI
#da população, seguindo distribuição normal, com valor esperado 100 e
#desvio padrão 15
import numpy as np
import matplotlib.pyplot as plt

mu = 100 # valor esperado
sigma = 15 #desvio padrão

#gerando 10000 valores aleatórios seguindo distribuição normal com
#valor esperado mu e desvio sigma
x = mu + sigma * np.random.randn(10000)  #amostra com 10000 valores aleatórios

n, bins, patches = plt.hist(x, 50)

plt.xlabel('Inteligencia')
plt.ylabel('Probabilidade')
plt.title('Histograma de QI')

plt.grid(True)
plt.show()
```
